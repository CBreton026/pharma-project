```python
df2 = df_fr_elect_prod_data.dropna(axis=1)
```

```python
#Filter rows 1 to 24 (0-based indexing for rows 0-23)
df_day1 = df2.iloc[0:24,2:]
# Calculate the column sums for rows 1 to 24 
sums_day1 = df_day1.sum(axis=0) 
# Calculate the percentage relative to column sums 
percentage1 = df_day1.div(sums_day1, axis=1) 
#Filter rows 1 to 24 (0-based indexing for rows 0-23)
df_day2 = df2.iloc[24:48,2:] 
# Calculate the column sums for rows 1 to 24 
sums_day2 = df_day2.sum(axis=0) 
# Calculate the percentage relative to column sums 
percentage2 = df_day2.div(sums_day2, axis=1) 
```
